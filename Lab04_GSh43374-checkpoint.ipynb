## 1.1) Importing the libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from xgboost import XGBRegressor
from sklearn.feature_selection import SelectFromModel
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor

from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt 
```

## 1.2) Reading the data

```python
data = pd.read_csv('Lab02_GSh43374_prepared.csv')
data.head()
```

## 1.3) Splitting the data

```python
features = data.drop('enrollment', axis=1) 
target = data['enrollment'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.25, random_state=42)
X_train.head()
```

## 1.4) Initiate some lists to store your experimentsâ€™ result

```python
models =[]
r2_list = []
rmse_score = []
modelNames = []
```

## 1.5) Selecting the feature transformation, scaling method along with random forest classifier

```python
# constructing the pipeline

# Polynomial feature transformation is used with degree 2
poly_features = PolynomialFeatures(degree=2, include_bias=False, interaction_only=False)

# robust scaling method is used
robustScaler = RobustScaler()

# Random Forest feature selection method is used using select from model func
selRandom = SelectFromModel(RandomForestClassifier(n_estimators = 100))
```

## 1.6) Constructing the pipeline

```python
# list of the classifier names is mentioned here that were used.
modelNames = ["Linear Regression", "Decision Tree Regression", "Gradient Boosting Regression", 
         "Random Forest Regression", "Ada Boost Regression", "XGB Regression", "LGBM Regressor",
        "CatBoost Regression"]

# list of pipe
pipes = [] 

# classifiers that are used.
clfs = [
    LinearRegression(),
    DecisionTreeRegressor(max_depth=3),
    GradientBoostingRegressor(),
    RandomForestRegressor(),
    AdaBoostRegressor(),
    XGBRegressor(),
    LGBMRegressor(),
    CatBoostRegressor()
]

# here first of all, the polynomial feature is applied and then robust scaling method is used followed by select random feature
# on random forest classifier. All of the transfomations are done and then different regressors are used to fit the model. 
for clf in clfs:
    pipe = Pipeline(steps=[("poly_feat", poly_features),
        ("scaler", robustScaler),("selectRandomFeature",selRandom),('classifier', clf)
    ])
    pipe.fit(X_train, y_train)
    pipes.append(pipe) 
    print(clf)
    y_pred = pipe.predict(X_test)
    r2_list.append(pipe.score(X_test, y_test))
    rmse_score.append(mean_squared_error(y_test, y_pred)**0.5)
    
# create dataframe of result
result = pd.DataFrame(zip(modelNames,r2_list,rmse_score), columns=['Regressor', 'R2', 'RMSE'])
```

## 1.7) Display the result

```python
result
```

## 1.8) Sorting the result by RMSE

```python
result.sort_values(by="RMSE")
```

From above outcomes, model CatBoost Regressor provided the least root mean square error as 4435.255773 and higher R squared as 0.754265. So, this model is considered as best model for me.
